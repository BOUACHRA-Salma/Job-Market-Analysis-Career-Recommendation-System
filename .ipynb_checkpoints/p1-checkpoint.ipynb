```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("job_titles_classification_extended.csv")
df.head()

df['skills_required'] = df['skills_required'].fillna('') #checks if there is any NaN and replace it with " "
tfidf = TfidfVectorizer(stop_words = None) #creating of an instance, stop_words tells the vectorizer to ignore common words like "the", "and" ..and keep onky the skills
skills_matrix = tfidf.fit_transform(df['skills_required']) # the vectorizer sacn the skills_required and store them and calculate the specific TF-IDF wight for every skill
#transform : converts the text into a matrix based on the learend weights
print(f"Shape of Skills Matrix: {skills_matrix.shape}")

# 1. Clean the string column: Remove '$' and 'k' symbols, and fill NaNs with a safe value
df['salary_cleaned'] = df['salary_range'].str.replace('$', '', regex=False).str.replace('k', '', regex=False).fillna('0-0')

# 2. Split the range into min and max salary columns (still in thousands)
df[['salary_min_k', 'salary_max_k']] = df['salary_cleaned'].str.split('-', expand=True).astype(float)

# 3. Calculate the average salary and multiply by 1000 to get the final numerical value
df['salary_avg'] = ((df['salary_min_k'] + df['salary_max_k']) / 2) * 1000

# Optional: Drop the intermediate columns
df = df.drop(columns=['salary_cleaned', 'salary_min_k', 'salary_max_k'])

# Result Check: df['salary_avg'] is now a clean numerical column
df.head()

# 1. Binary conversion for 'remote' (Yes=1, No=0)
df['is_remote'] = df['remote'].map({'Yes': 1, 'No': 0})

# 2. Text Normalization (Important for consistency and grouping)
df['job_title'] = df['job_title'].str.lower()
df['category'] = df['category'].str.lower()
df['location'] = df['location'].str.lower()
df.head()


#------------------------VISUALISATION----------------------


# =========================================================
# --- CONFIGURATION DU STYLE ---
# =========================================================

# sns.set_style() : Fonction de Seaborn qui dfinit l'apparence gnrale des graphiques.
# 'whitegrid' : Applique un arrire-plan blanc avec des lignes de grille horizontales,
#               ce qui est trs courant pour les graphiques statistiques.
sns.set_style("whitegrid")

# plt.rcParams[] : Dictionnaire de Matplotlib pour modifier les paramtres globaux.
# 'figure.figsize' : Dfinit la taille par dfaut de toutes les figures (graphiques) cres par la suite.
# [12, 6] : Fixe la taille  12 units de largeur par 6 units de hauteur.
plt.rcParams['figure.figsize'] = [12, 6] 

# =========================================================
# 1. Distribution Gnrale des Salaires (Histogramme)
# =========================================================

# plt.figure() : Cre une nouvelle figure (une zone de dessin vide) avec une taille spcifique.
# (10, 6) : La taille de cette figure est dfinie  10 de large et 6 de haut.
plt.figure(figsize=(10, 6))

# sns.histplot() : Fonction de Seaborn qui cre un histogramme (et une estimation de densit de noyau).
# df['salary_avg'] : La colonne de donnes numriques utilise pour le graphique.
# bins=30 : Divise la plage de donnes en 30 intervalles (barres).
# kde=True : Ajoute une courbe de densit (Kernel Density Estimate) pour lisser la distribution.
# color='skyblue' : Dfinit la couleur des barres et de la courbe.
sns.histplot(df['salary_avg'], bins=30, kde=True, color='skyblue')

# plt.title() : Ajoute un titre au graphique.
# fontsize=16 : Dfinit la taille de la police du titre.
plt.title('Distribution des Salaires Annuels\n', fontsize=16)

# plt.xlabel() / plt.ylabel() : Ajoutent des tiquettes aux axes X et Y, respectivement.
plt.xlabel('Salaire Annuel Moyen', fontsize=12)
plt.ylabel('Frquence des Offres', fontsize=12)

# plt.ticklabel_format() : Contrle la mise en forme des tiquettes des axes.
# style='plain' : Force l'affichage en nombre normal (pas de notation scientifique comme 1e+06).
# axis='x', useOffset=False : Applique ceci uniquement  l'axe X.
plt.ticklabel_format(style='plain', axis='x', useOffset=False)

# plt.show() : Affiche la figure Matplotlib qui vient d'tre cre.
plt.show()

# print("\n\n") : Ajoute deux lignes vides dans la console pour sparer les sorties.
print("\n\n") 


# =========================================================
# 2. Comparaison des Salaires par Catgorie (Box Plot)
# =========================================================

# plt.figure() : Cre la deuxime figure.
plt.figure(figsize=(14, 7))

# sns.boxplot() : Fonction de Seaborn qui affiche la distribution des donnes via des botes.
# x='category', y='salary_avg' : Dfinit la variable catgorielle (X) et la variable numrique (Y).
# data=df : Spcifie le DataFrame  utiliser.
# hue='category' : Utilise la mme colonne 'category' pour attribuer des couleurs diffrentes  chaque bote (meilleure pratique Seaborn).
# palette='viridis' : Dfinit la palette de couleurs  utiliser.
# legend=False : Supprime la lgende automatique, car les couleurs sont videntes.
sns.boxplot(
    x='category', 
    y='salary_avg', 
    data=df, 
    hue='category', 
    palette='viridis',
    legend=False 
)

# plt.title() : Ajout du titre du Box Plot.
plt.title('Salaires Annuels par Catgorie d\'Emploi\n', fontsize=16)
plt.xlabel('Catgorie', fontsize=12)
plt.ylabel('Salaire (USD)', fontsize=12)

# plt.xticks() : Personnalise les graduations de l'axe X.
# rotation=45 : Fait pivoter les noms des catgories de 45 degrs pour viter qu'ils ne se chevauchent.
plt.xticks(rotation=45) 

# plt.show() : Affiche la deuxime figure.
plt.show()

print("\n\n")

# =========================================================
# 3. Top Salaires Mdians par Catgorie (Bar Plot)
# =========================================================

# df.groupby() : Fonction de Pandas qui regroupe le DataFrame par la colonne spcifie ('category').
# ['salary_avg'].median() : Calcule la mdiane de 'salary_avg' pour chaque groupe ('category').
# .sort_values(ascending=False) : Trie les rsultats du plus grand au plus petit.
# .reset_index() : Transforme la srie rsultante en un DataFrame standard (avec 'category' comme colonne).
median_salaries_category = df.groupby('category')['salary_avg'].median().sort_values(ascending=False).reset_index()

# .columns = [...] : Renomme les colonnes du DataFrame group pour tre explicite.
median_salaries_category.columns = ['category', 'salary_avg'] 

# plt.figure() : Cre la troisime figure.
plt.figure(figsize=(10, 6))

# sns.barplot() : Fonction de Seaborn qui trace une barre pour la moyenne (ou, ici, la mdiane) de chaque catgorie.
# x='category', y='salary_avg' : Axes dfinis par le nouveau DataFrame group.
# data=median_salaries_category : Utilise le DataFrame des mdianes calcules.
# palette='Spectral' : Utilise une palette de couleurs diffrente.
sns.barplot(
    x='category', 
    y='salary_avg', 
    data=median_salaries_category, 
    hue='category', 
    palette='Spectral',
    legend=False
)

plt.title('Salaire Mdian le Plus lev par Catgorie\n', fontsize=16)
plt.xlabel('Catgorie', fontsize=12)
plt.ylabel('Salaire Mdian Annuel (USD)', fontsize=12)

# ha='right' : Alignement horizontal des tiquettes sur la droite aprs la rotation.
plt.xticks(rotation=45, ha='right')

# plt.ticklabel_format() : Assure un affichage numrique normal pour l'axe Y.
plt.ticklabel_format(style='plain', axis='y', useOffset=False)

# for index, row in median_salaries_category.iterrows() : Boucle de Pandas qui itre sur chaque ligne du DataFrame group.
for index, row in median_salaries_category.iterrows():
    # plt.text() : Ajoute du texte sur le graphique  des coordonnes spcifiques.
    # index : Position X (la position de la barre).
    # row['salary_avg'] + 5000 : Position Y (juste au-dessus de la barre).
    # f"{...}" : Formatage de la chane de caractres (f-string) pour afficher la valeur avec sparateur de milliers et "DH".
    # ha="center" : Centre le texte au milieu de la barre.
    plt.text(index, row['salary_avg'] + 5000, 
             f"{row['salary_avg']:,.0f} DH", 
             color='black', ha="center", fontsize=10)
    
# plt.show() : Affiche la troisime figure.
plt.show()

print("\n\n")

# =========================================================
# 4. Top 5 des Salaires Mdians par Localisation (Bar Plot)
# =========================================================

# df[...] : Cre un sous-ensemble du DataFrame en filtrant les lignes.
# df['location'] != 'other' : Inclut uniquement les lignes o 'location' n'est PAS la catgorie 'other'.
df_specific_locations = df[df['location'] != 'other'] 

# Calculer la mdiane pour les 5 localisations principales (mmes fonctions de Pandas que prcdemment).
# .head(5) : Limite le rsultat aux 5 premires lignes (les 5 plus grandes mdianes).
median_salaries_location = df_specific_locations.groupby('location')['salary_avg'].median().sort_values(ascending=False).head(5).reset_index()

# plt.figure() : Cre la quatrime figure.
plt.figure(figsize=(10, 6))

# sns.barplot() : Trace les barres pour le Top 5 des localisations.
# palette='rocket' : Utilise une autre palette squentielle (couleurs chaudes).
sns.barplot(
    x='location', 
    y='salary_avg', 
    data=median_salaries_location, 
    hue='location', 
    palette='rocket',
    legend=False
)

plt.title('Top 5 des Salaires Mdians par Localisation\n', fontsize=16)
plt.xlabel('Localisation', fontsize=12)
plt.ylabel('Salaire Mdian Annuel (USD)', fontsize=12)
plt.xticks(rotation=0)
plt.ticklabel_format(style='plain', axis='y', useOffset=False)

# Afficher les valeurs exactes (mme logique que l'tape 3).
for index, row in median_salaries_location.iterrows():
    # Ajustement de l'offset Y et formatage DH.
    plt.text(index, row['salary_avg'] + 5000, 
             f"{row['salary_avg']:,.0f}DH", 
             color='black', ha="center", fontsize=10)

# plt.show() : Affiche la dernire figure.
plt.show()
print('\n\n')



#    Frquence Comptence vs. Titre de Poste
# (On rexcute ici la normalisation pour la robustesse)
df['job_title'] = df['job_title'].str.lower().str.strip()
df['skills_cleaned'] = df['skills_required'].str.lower().str.replace(r',\s*', ',', regex=True).fillna('')


# --- 1. IDENTIFIER LES TOP JOBS et FILTRER ---

# Compter la frquence de chaque job_title
top_jobs = df['job_title'].value_counts().head(10).index.tolist()

# Filtrer le DataFrame pour ne garder que ces 10 jobs
df_filtered = df[df['job_title'].isin(top_jobs)].copy()


# --- 2. TRANSFORMATION DES COMPTENCES (MATRICE DE COMPTAGE) ---

# Le CountVectorizer est parfait pour obtenir des comptes de frquences.
# tokenizer : Utilise la virgule comme sparateur pour chaque comptence.
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(',')) 
skill_counts_matrix = vectorizer.fit_transform(df_filtered['skills_cleaned'])

# Crer un DataFrame des comptes de comptences
skill_df = pd.DataFrame(skill_counts_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=df_filtered.index)

# Joindre le titre du poste au DataFrame des comptences
skill_df['job_title'] = df_filtered['job_title']


# --- 3. AGGREGATION ET VISUALISATION (HEATMAP) ---

# Calculer la somme des occurrences de chaque comptence par titre de poste
job_skill_pivot = skill_df.drop(columns=['job_title']).groupby(skill_df['job_title']).sum()

# Filtrer pour les 15-20 comptences les plus frquentes pour un graphique lisible
top_skills_overall = job_skill_pivot.sum(axis=0).sort_values(ascending=False).head(20).index
job_skill_heatmap_data = job_skill_pivot[top_skills_overall]

plt.figure(figsize=(16, 10))
# sns.heatmap() : Affiche la matrice de co-occurrence avec une chelle de couleurs.
sns.heatmap(
    job_skill_heatmap_data, 
    annot=True,              # Affiche le nombre exact dans chaque cellule
    cmap='YlGnBu',           # Palette de couleurs (jaune-vert-bleu)
    fmt='d',                 # Format des nombres (entiers)
    linewidths=.5,           # Lignes fines entre les cellules
    cbar_kws={'label': 'Nombre d\'Offres Requrant la Comptence'}
)
plt.title('Frquence des Comptences par Titre de Poste (Top 10 Jobs)', fontsize=18)
plt.xlabel('Comptences', fontsize=14)
plt.ylabel('Titre de Poste', fontsize=14)
plt.xticks(rotation=60, ha='right')
plt.show()

```
