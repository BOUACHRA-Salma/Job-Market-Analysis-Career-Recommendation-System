<h2><b><u>Importation des bibliothèques nécessaires : </u></b></h2>

```python
import pandas as pd        # Bibliothèque utilisée pour manipuler des données tabulaires (DataFrames)
import numpy as np         # Bibliothèque pour le calcul numérique et les tableaux multidimensionnels

from sklearn.feature_extraction.text import TfidfVectorizer  
# Outil de scikit-learn qui transforme du texte en vecteurs numériques basés sur la méthode TF-IDF
# (permet de mesurer l'importance d'un mot dans un document)

from sklearn.cluster import KMeans  
# Algorithme de clustering non supervisé permettant de regrouper les données en K groupes selon leurs similarités

from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt  
# Bibliothèque de visualisation pour créer des graphiques (courbes, barres, camemberts, etc.)

import seaborn as sns      
# Bibliothèque basée sur Matplotlib offrant des graphes plus esthétiques et simples à utiliser
```

<h2><u><b>Collecte de données : </b></u></h2>

```python
# Chargement des données depuis un fichier CSV nommé "job_titles_classification_extended.csv"
df = pd.read_csv("job_titles_classification_extended.csv")

# Affiche les 5 premières lignes du DataFrame pour visualiser rapidement la structure des données
df.head()
```

<h2><u><b>Prétraitement des données : </b></u></h2>

```python
df['skills_required'] = df['skills_required'].fillna('')
# Remplace toutes les valeurs manquantes (NaN) dans la colonne "skills_required" par une chaîne vide.
# Cela évite les erreurs lorsque le texte sera transformé en vecteurs TF-IDF.

tfidf = TfidfVectorizer(stop_words=None)
# Création d’une instance du vecteuriseur TF-IDF.
# Le paramètre stop_words=None signifie que l'on ne retire aucun mot fréquent.
# Tous les mots (donc toutes les compétences listées) seront pris en compte.

skills_matrix = tfidf.fit_transform(df['skills_required'])
# Le vecteuriseur apprend d'abord le vocabulaire (fit) à partir de la colonne "skills_required".
# Puis il transforme chaque ligne en un vecteur numérique basé sur les poids TF-IDF (transform).
# Le résultat est une matrice sparse où chaque ligne représente un job et chaque colonne une compétence.

print(f"Shape of Skills Matrix: {skills_matrix.shape}")
# Affiche la taille de la matrice TF-IDF : 
# (nombre de lignes = nombre d’emplois, nombre de colonnes = nombre de compétences uniques détectées)
```

```python
# 1. Nettoyer la colonne de chaîne : retirer les symboles '$' et 'k', et remplacer les NaN par une valeur sûre
df['salary_cleaned'] = (
    df['salary_range']
      .str.replace('$', '', regex=False)   # enlève le signe '$' si présent
      .str.replace('k', '', regex=False)   # enlève le 'k' signifiant « milliers »
      .fillna('0-0')                       # si valeur manquante, remplace par '0-0'
)

# 2. Séparer la plage en deux colonnes : salaire minimum et maximum (toujours en milliers)
df[['salary_min_k', 'salary_max_k']] = (
    df['salary_cleaned']
      .str.split('-', expand=True)    # sépare la chaîne autour du tiret '-'
      .astype(float)                  # convertit les chaînes résultantes en nombres (float)
)

# 3. Calculer le salaire moyen, puis multiplier par 1000 pour obtenir une valeur numérique finale
df['salary_avg'] = (
    (df['salary_min_k'] + df['salary_max_k']) / 2   # calcul de la moyenne des deux bornes
) * 1000                                            # conversion de « milliers » en unité monétaire

# Optionnel : Supprimer les colonnes intermédiaires, jugées inutiles après calcul
df = df.drop(columns=['salary_cleaned', 'salary_min_k', 'salary_max_k'])

# Vérification : df['salary_avg'] est désormais une colonne numérique « propre »
df.head()
```

```python
# 1. Conversion binaire pour la colonne 'remote' (Yes = 1, No = 0)
df['is_remote'] = df['remote'].map({'Yes': 1, 'No': 0})
# La méthode map() remplace chaque valeur par un équivalent numérique.
# Cela facilite les analyses statistiques et l'entraînement de modèles ML.

# 2. Normalisation du texte (important pour la cohérence et le regroupement)
df['job_title'] = df['job_title'].str.lower()
# Convertit tous les intitulés de poste en minuscules pour éviter que 'Data Scientist'
# et 'data scientist' soient traités comme différents.

df['category'] = df['category'].str.lower()
# Même logique : uniformiser les catégories en minuscules.

df['location'] = df['location'].str.lower()
# Rend toutes les localisations cohérentes (ex. 'New York' et 'new york').

df.head()
# Affiche les premières lignes pour vérifier les transformations.
```

```python
# Compter le nombre d'occurrences pour chaque catégorie,
# puis trier automatiquement du plus fréquent au moins fréquent
category_counts = df['category'].value_counts()

# Affiche les résultats : chaque catégorie avec son nombre d'apparitions
print(category_counts)
```

<h2><u><b>Exploration des données : </b></u></h2>

```python
# Crée une nouvelle figure (zone de dessin) avec une taille de 12 pouces par 7
plt.figure(figsize=(12, 7))

# Normalise les catégories en minuscules (pour éviter les doublons comme "Tech" et "tech")
df['category'] = df['category'].str.lower()

# Crée un boxplot (diagramme en boîte) pour visualiser la distribution des salaires par catégorie
sns.boxplot(
    x='category',         # Axe horizontal : les catégories d'emplois
    y='salary_avg',       # Axe vertical : le salaire moyen
    data=df,              # Le DataFrame contenant les données
    hue='category',       # Permet de colorer chaque catégorie différemment
    palette='viridis',    # Palette de couleurs utilisée
    legend=False          # Masque la légende (inutile car chaque catégorie est déjà affichée sur l'axe X)
)

# Titre du graphique
plt.title('Average Salary Distribution by Job Category')

# Nom de l'axe X
plt.xlabel('Job Category')

# Nom de l'axe Y
plt.ylabel('Average Salary (USD)')

# Affiche le graphique
plt.show()
```

```python
# Crée une nouvelle figure (zone de dessin) avec une taille de 10x6 pouces
plt.figure(figsize=(10, 6))

# Définit un ordre logique pour les niveaux d'expérience
experience_order = ['entry', 'mid', 'senior', 'lead']

# S'assure que la colonne 'experience_level' respecte cet ordre
df['experience_level'] = pd.Categorical(
    df['experience_level'].str.lower(),  # Convertit en minuscules pour uniformiser
    categories=experience_order,         # Spécifie l'ordre voulu
    ordered=True                         # Indique qu'il s'agit d'une catégorie ordonnée
)

# Crée un graphique de comptage (countplot)
# 'hue' sépare les barres en fonction du statut remote (Yes/No)
sns.countplot(
    x='experience_level',                # Axe X : niveaux d'expérience
    hue='remote',                        # Couleur des barres selon le travail à distance
    data=df,                             # Les données
    order=experience_order,              # Utilise l'ordre défini plus haut
    palette={'Yes': 'hotpink', 'No': 'crimson'}  # Couleurs personnalisées
)

# Titre du graphique
plt.title('Remote Work Availability by Experience Level')

# Nom de l'axe X
plt.xlabel('Experience Level')

# Nom de l'axe Y
plt.ylabel('Number of Listings')

# Légende précisant ce que signifient les couleurs
plt.legend(title='Remote Offered')

# Ajoute une grille horizontale pour faciliter la lecture
plt.grid(axis='y', linestyle='--')

# Affiche le graphique
plt.show()
```

<h2><u><b>Modélisation / Analyse : </b></u></h2>

```python
k = 10  # Nombre de clusters (à ajuster en fonction de vos données)
kmeans = KMeans(n_clusters=k, random_state=42)  # Initialisation du modèle KMeans avec k clusters et une graine aléatoire fixe pour la reproductibilité

labels = kmeans.fit_predict(skills_matrix)  # Entraînement du modèle KMeans sur la matrice des compétences et attribution d'un cluster à chaque échantillon
```

```python
df['cluster'] = labels
# pour associer à chaque ligne de ton DataFrame le numéro du cluster auquel elle appartient.
```

```python
for i in range(k):
    print(f"\nCluster {i} sample job titles:")  # Affiche le numéro du cluster courant

    # Filtre le DataFrame pour ne garder que les lignes appartenant au cluster i,
    # puis compte la fréquence des différents intitulés de poste (job_title) dans ce cluster,
    # et affiche les 10 intitulés les plus fréquents
    print(df[df['cluster'] == i]['job_title'].value_counts().head(10))
```

```python
# pour calculer la qualite du clustering
score = silhouette_score(skills_matrix, labels)
print(f'Silhouette score: {score:.3f}')
```
